```
import numpy as np
from PIL import Image, ImageDraw
import face_recognition
```

```
im1 = face_recognition.load_image_file('atatürk1.jpg')
im2 = face_recognition.load_image_file('Lenna_(test_image).png')
im3 = face_recognition.load_image_file('240full-lena-söderberg.jpg')
im4 = face_recognition.load_image_file('cagatay_odabasi-248x300.png')
im5 = face_recognition.load_image_file('66dd5b07a6-avatar-orig.jpg')
```

```
face_locations1 = face_recognition.face_locations(im1)
face_locations2 = face_recognition.face_locations(im2)
face_locations3 = face_recognition.face_locations(im3)
face_locations4 = face_recognition.face_locations(im4)
face_locations5 = face_recognition.face_locations(im5)
```

```
e1 = face_recognition.face_encodings(im1)[0]
e2 = face_recognition.face_encodings(im2)[0]
e3 = face_recognition.face_encodings(im3)[0]
e4 = face_recognition.face_encodings(im4)[0]
e5 = face_recognition.face_encodings(im5)[0]
```

```
result1 = face_recognition.compare_faces([e1],e2)
result2 = face_recognition.compare_faces([e2],e3)
result3 = face_recognition.compare_faces([e3],e4)
result4 = face_recognition.compare_faces([e4],e1)
result5 = face_recognition.compare_faces([e5],e2)

print(result1)
print(result2)
print(result3)
print(result4)
print(result5)
```

```
[top1,right1,bottom1,left1] = face_locations1[0]
[top2,right2,bottom2,left2] = face_locations2[0]
[top3,right3,bottom3,left3] = face_locations3[0]
[top4,right4,bottom4,left4] = face_locations4[0]
[top5,right5,bottom5,left5] = face_locations5[0]
```
